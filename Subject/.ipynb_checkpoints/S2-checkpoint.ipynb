## Quel est l'impact du stress sur les performances scolaires ? 

## ⚙️ Régression linéaire, Corrélation et Test

### 1-Importation et aperçu de notre dataset

```python
import pandas as pd

data = pd.read_csv("Data/StressLevelDataset_S2.csv")
data.head()
```

### 2-Valeurs manquantes et nettoyage

```python
print(data.describe().transpose())
```

```python
print(data.isna().sum())
```

### 3-Corrélation ( Pearson )

```python
from scipy.stats import pearsonr

r, p = pearsonr(data["stress_level"], data["academic_performance"])
print(f"Corrélation r = {r:.3f}, p-value = {p:.3f}")
```

### 4-Modèle de régression de linéaire

```python
import statsmodels.api as sm

X = sm.add_constant(data["stress_level"])
model = sm.OLS(data["academic_performance"], X).fit()
print(model.summary())
```

### 5-Résumé

Les données montrent une relation négative forte entre le stress et la performance scolaire (r = –0.721, p < 0.001).
Le modèle de régression confirme que le stress réduit significativement la performance : chaque unité de stress diminue le score académique d’environ 1.24 points (β = –1.24).
Avec un R² de 0.520, le stress explique plus de la moitié de la variance des performances, ce qui en fait un facteur majeur.

Une nuance importante : dans la littérature, un stress léger peut parfois agir comme un stimulant, il favorise donc l’attention, le gain d'énergie et la mobilisation cognitive.

Mais dans ce dataset, cette zone “optimale” n’apparaît pas : la relation est globalement et fortement négative, ce qui suggère que les niveaux de stress observés dépassent probablement le seuil où il devient bénéfique.

Conclusion : Le stress n’est pas seulement un inconfort, c’est un frein mesurable à la performance scolaire. Même si un stress léger peut théoriquement être stimulant, les données montrent que dans la majorité des cas, il dégrade la performance de manière significative.
